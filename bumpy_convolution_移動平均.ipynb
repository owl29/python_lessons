```python
# numpy.convolutionを使った移動平均でのノイズフィルター
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 500)
y1 = np.sin(x) # まずは元の信号を
y2 = y1 + np.random.randn(500)*0.3 # ノイズを混ぜる

v = np.ones(10)/10.0 # 移動平均をとるための配列vを設定。今回は前後５つの値を用いて平均をとる。
y3 = np.convolve(y2, v, mode='same')  # グラフを描く都合上'same'で。

plt.plot(x, y2,'b',linewidth=1)
plt.plot(x, y3,'r',linewidth=2)
plt.plot(x, y1,'y',linewidth=3)


plt.show()
```
